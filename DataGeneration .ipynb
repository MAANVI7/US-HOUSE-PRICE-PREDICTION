```python
#Imporing CASE-SHILLER Index data 

df_CSI = pd.read_csv("CSUSHPISA.csv")

#Changing dtype of date column
df_CSI["DATE"] = pd.to_datetime(df_CSI["DATE"])

#Selecting data post JUNE 2002 because last 20 years data is needed
A = df_CSI["DATE"] >= "2002-07-01"
df_CSI = df_CSI[A]

#Resetting Index
df_CSI.reset_index(inplace = True)
df_CSI.drop(columns = ["index"], inplace = True)

# Creating "Year" and "Month" columns
df_CSI["Year"] = df_CSI["DATE"].dt.year
df_CSI["Month"] = df_CSI["DATE"].dt.month
print(df_CSI.shape)
df_CSI.tail()
```
